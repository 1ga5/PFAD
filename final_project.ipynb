# Подготовка данных

```python
# Замена названий столбцов
df = df.set_axis(['номер', 'канал продаж', 'продуктовый сегмент', 'менеджер',
       'название', 'продавец', 'город', 'продажи факт',
       'продажи план', 'дата сделки', 'дата оплаты',
       'дней между', 'статус'], axis=1)

print(df)
```

```python
print(df.isnull().sum())
print(df.duplicated().sum())
print(df[df["дата оплаты"].isna()])
```

```python
df.dropna(subset=['дата оплаты'], inplace=True)
print(df.isnull().sum())
```

```python
df["дней между"] = pd.to_numeric(df["дней между"].astype("int"))
```
